```python
IndexedRegTeamTotals = RegTeamTotals.set_index(['Year'])
IndexedRegTeamTotals = IndexedRegTeamTotals.drop(columns=['Unnamed: 8','Unnamed: 11','Unnamed: 14','Unnamed: 17', 'Unnamed: 20'])
IndexedRegTeamTotals = IndexedRegTeamTotals.rename(columns={'W.1': 'Conf Wins', 'L.1':'Conf Losses', 'W.2':'Home Wins', 'L.2':'Home Losses', 'W.3':'Away Wins', 'L.3':'Away Losses'})
TourneyTeamTotals = IndexedRegTeamTotals[IndexedRegTeamTotals['School'].str.contains('NCAA')]
NonTourneyTeamTotals = IndexedRegTeamTotals[IndexedRegTeamTotals['School'].apply(lambda x:'NCAA' not in x)]
SeasonTourneyTeams= TourneyTeamTotals.groupby(TourneyTeamTotals.index)
```

```python
sns.set_style(style="whitegrid")

# Selecting columns for plots
tourney_cols = TourneyTeamTotals.select_dtypes(include=['int64', 'float']).columns
nontourney_cols = NonTourneyTeamTotals.select_dtypes(include=['int64', 'float']).columns


# Plotting distributions of numerical features
plt.figure(figsize=(25, 20))
for i, col in enumerate(tourney_cols):
    plt.subplot(8, 4, i + 1)
    sns.histplot(TourneyTeamTotals[col], kde=True)
    plt.title(col)
    plt.tight_layout()
for i, col in enumerate(nontourney_cols):
    plt.subplot(8, 4, i + 1)
    sns.histplot(NonTourneyTeamTotals[col], kde=True)
    plt.title(col)
    plt.tight_layout()

plt.show()
```
